```python
import numpy as np
import matplotlib as plt
import pylab as pl
%pylab inline
%matplotlib inline
np.random.seed(5)
```

```python
mean = 10
sample_size = np.random.randint(1,high=2000,size=100)
print(sample_size.shape)
print (sample_size)
```

```python
md = {}

md['chisq'] = {} 
md['normal'] = {}
md['poisson'] = {}
md['binomial'] = {}
md['logistic'] = {}

for n in sample_size:
    md['chisq'][n] = np.random.chisquare(mean, size = n)

for n in sample_size:
    md['normal'][n] = np.random.randn(n) + mean

for n in sample_size:
    md['poisson'][n] = np.random.poisson(mean, size = n)

for n in sample_size:
    md['binomial'][n] = np.random.binomial(50,0.2, size = n)

for n in sample_size:
    md['logistic'][n] = np.random.logistic(mean, size = n)
```

```python
md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        #and plot it
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(sample_size), max(sample_size)], [mean, mean], 'k')

```

Figure 1. The larger the sample size is, the closer the mean of the sample is to the mean of the population. This trend reflects the Central Limit Theorem.

```python
allmeans = []
allmeans += (list(md['chisq']['means'].values()))
allmeans += (list(md['normal']['means'].values()))
allmeans += (list(md['poisson']['means'].values()))
allmeans += (list(md['binomial']['means'].values()))
allmeans += (list(md['logistic']['means'].values()))
```

```python
pl.figure(figsize=(12, 12))
pl.hist(allmeans, bins=24)
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.title("Distribution of the Samples' Mean", fontsize = 22)
```

Figure 6. Most of the samples' mean are close to the mean of the population (population mean = 10).

```python
#Extra Credit
allmeans = np.asarray(allmeans)
allmeans.sort()
```

```python
from scipy.stats import norm
pl.figure(figsize=(12, 12))
fit = norm.pdf(allmeans, np.mean(allmeans), np.std(allmeans))
pl.plot(allmeans,fit,'y',)
pl.hist(allmeans, density=True,bins=24)
pl.title("Distribution of the Samples' Mean with Gaussian", fontsize = 22)
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

Figure 7. Fitted a line of Gaussian distribution to the histogram of the distribution of sample's mean.
